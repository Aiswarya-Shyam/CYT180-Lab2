### What `.describe()` tells us

**describe()** provides:
1) **count** – number of non-null values
2) **mean** – the average value
3) **std** – standard deviation, shows how spread out values are
4) **min/max** – smallest and largest values in the column
5) **25%/50%/75%** – quartiles that divide data into four equal parts; the 50% is the median

## Task 3: Data Filtering (Identify High-Risk Cities)

```python
# Name: Aiswarya Shyam Sundar
# Student ID: 111172268


# return a pandas series of (True/False) values
# same length as the DataFrame of whichever rows meet the condition
high_risk = df[df['failed_logins'] > 50]
high_risk
```

```python
# Name: Aiswarya Shyam Sundar
# Student ID: 111172268

# Step 2 – Sort the filtered results by failed_logins, highest first
high_risk_sorted = high_risk.sort_values('failed_logins', ascending=False)
high_risk_sorted
```

```python
# Name: Aiswarya Shyam Sundar
# Student ID: 111172268

# Step 3 – Count how many high-risk cities there are
num_high_risk = len(high_risk_sorted)
print(f"Number of high-risk cities (failed_logins > 50): {num_high_risk}")
```

**Why sort and count?**
Sorting brings the high risk cities to the top so that the worst problems can be seen immediately without scanning every row. Counting tells us how common the problem is.

**How does this help an analyst?**
Time is a constraint for analysts. They can triage quickly (starting with the highest risk cities first) and not treat all cities equally by sorting by failed logins. This is risk based prioritisation as used in real SOCs.

## Task 4: Simple Analysis of Cities by State

```python
# Name: Aiswarya Shyam Sundar
# Student ID: 111172268

# Step 1 – Count how many cities exist in each state
city_counts_by_state = df['State'].value_counts()
print("Number of cities per state:")
print(city_counts_by_state)
```

```python
# Name: Aiswarya Shyam Sundar
# Student ID: 111172268

# Step 2 – Identify the state with the highest number of cities
most_cities_state = city_counts_by_state.idxmax()
print(f"State with the most cities: {most_cities_state} ({city_counts_by_state[most_cities_state]} cities)")
```

```python
# Name: Aiswarya Shyam Sundar
# Student ID: 111172268

# Step 3 – Display the top 5 states with the most cities
print("Top 5 states by number of cities:")
print(city_counts_by_state.head(5))
```

**idxmax()** returns the label from the index that’s holding the biggest value in a Series. In this case it gives us the state code with the highest city count ,so we don’t need to go through and manually search all the other results one by one.

# Task 5: Visualization of Failed Logins

```python
# Name: Aiswarya Shyam Sundar # Student ID: 111172268
import matplotlib.pyplot as plt

# Combine City and State so each bar has a unique label (e.g., "Springfield, OH")
df['City_State'] = df['City'].str.strip() + ', ' + df['State'].str.strip()
colors = ['red' if val > 50 else 'steelblue' for val in df['failed_logins']]

plt.figure(figsize=(20, 6))
plt.bar(df['City_State'], df['failed_logins'], color=colors)

plt.xticks(rotation=90, fontsize=7)
plt.xlabel('City, State', fontsize=12)
plt.ylabel('Failed Logins', fontsize=12)
plt.title('Failed Logins per City\n(Red = High-Risk: > 50 failed logins)', fontsize=14)
plt.axhline(y=50, color='orange', linestyle='--', linewidth=1.5, label='Threshold (50)')
plt.legend()

plt.tight_layout()
plt.show()
```

### Reflection – Task 5

**How could this visualization help a cybersecurity analyst respond faster?**  
The bar chart gives an instant visual overview of the entire dataset. Red bars immediately draw the eye to cities exceeding the threshold, no need to scroll through rows of numbers. In a live SOC dashboard, a similar chart would let an analyst spot anomalies in seconds and jump straight to investigating the flagged cities, reducing mean time to respond (MTTR).

## Task 6: Calculate Failed Login Rate

```python
# Name: Aiswarya Shyam Sundar
# Student ID: 111172268

# Step 1 – Create a new column 'failed_login_rate'
# Divide failed_logins by login_attempts for each row.
# The result is a proportion between 0 and 1
df['failed_login_rate'] = df['failed_logins'] / df['login_attempts']

# Step 2 – Display the DataFrame to confirm the new column was created
print("DataFrame with the new 'failed_login_rate' column:")
df.head(10)
```

```python
# Name: Aiswarya Shyam Sundar
# Student ID: 111172268

# Step 3 – Sort the DataFrame by failed_login_rate in descending order
df_sorted_rate = df.sort_values('failed_login_rate', ascending=False)

# Step 4 – Display the top 5 cities with the highest failed login rate
print("Top 5 cities by failed login rate:")
df_sorted_rate[['City', 'State', 'failed_logins', 'login_attempts', 'failed_login_rate']].head(5)
```

### Why is the failed login rate more informative than raw counts?

Raw numbers are not very meaningful, because cities have very different total traffic of logins. This is far less of a concern than a city that has 30 failed login attempts out of 35 attempts (86% rate) rather than 90 failed login attempts out of 1000 attempts (9% rate). The rate normalises the data by dividing failures by total attempts, enabling cities to be compared on an even playing field regardless of the number of overall login attempts. In cybersecurity, a high failure rate is a much stronger indicator of a brute-force attack or credential-stuffing campaign than a high failure count alone.
